# Parte 1

```python
import matplotlib.pyplot as plt  # Importando a biblioteca matplotlib para plotagem

# Criando uma figura com um tamanho específico (2 polegadas de largura, 5 polegadas de altura)
fig, ax = plt.subplots(figsize=(2, 5))  # Instanciando a figura e o eixo

# Pontos de dados consistindo de tuplas de valores (x, y)
dados = [(35, 42), (41, 45), (49, 33), (59, 62), (76, 75), (80, 96), (85, 91)]

# Anos correspondentes aos pontos de dados, representados como tuplas de strings
anos = [('2014', '2015'), ('2014', '2015'), ('2014', '2015'), ('2014', '2015'),
        ('2014', '2015'), ('2014', '2015'), ('2014', '2015')]

# Plotando cada ponto de dado com seu ano correspondente usando uma linha preta e marcadores circulares
for dado_, ano_ in zip(dados, anos):
    ax.plot(ano_, dado_, color='black', marker='o')

# Removendo as bordas direita, esquerda e superior do gráfico
ax.spines['right'].set_visible(False)  # Remover a borda direita
ax.spines['left'].set_visible(False)   # Remover a borda esquerda
ax.spines['top'].set_visible(False)    # Remover a borda superior

# Removendo os marcadores do eixo y
ax.get_yaxis().set_visible(False)  # Ocultar os marcadores do eixo y

# Definindo o rótulo para o eixo x como 'Ano da pesquisa'
ax.set_xlabel('Ano da pesquisa')  # Definir o rótulo do eixo x

# Exibindo o gráfico
plt.show()  # Mostrar o gráfico
```

```python
import matplotlib.pyplot as plt  # Importando a biblioteca matplotlib para plotagem

# Criando uma figura com um tamanho específico (2 polegadas de largura, 5 polegadas de altura)
fig, ax = plt.subplots(figsize=(2, 5))  # Instanciando a figura e o eixo

# Pontos de dados consistindo de tuplas de valores (x, y)
dados = [(35, 42), (41, 45), (49, 33), (59, 62), (76, 75), (80, 96), (85, 91)]

# Anos correspondentes aos pontos de dados, representados como tuplas de strings
anos = [('2014', '2015'), ('2014', '2015'), ('2014', '2015'), ('2014', '2015'),
        ('2014', '2015'), ('2014', '2015'), ('2014', '2015')]

# Plotando cada ponto de dado com seu ano correspondente usando uma linha preta e marcadores circulares
for dado_, ano_ in zip(dados, anos):
    ax.plot(ano_, dado_, color='black', marker='o')

# Removendo as bordas direita, esquerda e superior do gráfico
ax.spines['right'].set_visible(False)  # Remover a borda direita
ax.spines['left'].set_visible(False)   # Remover a borda esquerda
ax.spines['top'].set_visible(False)    # Remover a borda superior

# Removendo os marcadores do eixo y
ax.get_yaxis().set_visible(False)  # Ocultar os marcadores do eixo y

# Definindo o rótulo para o eixo x como 'Ano da pesquisa'
ax.set_xlabel('Ano da pesquisa')  # Definir o rótulo do eixo x

# Adicionando rótulos de porcentagem aos pontos de dados para cada ano
for dado_, ano_ in zip(dados, anos):
    ano_anterior, ano_posterior = ano_
    dado_anterior, dado_posterior = dado_

    # Adicionando texto com a porcentagem à esquerda e à direita de cada ponto de dado
    ax.text(ano_anterior, dado_anterior, f'{dado_anterior}%  ', ha='right')
    ax.text(ano_posterior, dado_posterior, f'  {dado_posterior}%', ha='left')

plt.show()  # Exibindo o gráfico
```

# Parte 3

```python
import matplotlib.pyplot as plt  # Importando a biblioteca matplotlib para plotagem

# Criando uma figura com um tamanho específico (2 polegadas de largura, 5 polegadas de altura)
fig, ax = plt.subplots(figsize=(2, 5))  # Instanciando a figura e o eixo

# Pontos de dados consistindo de tuplas de valores (x, y)
dados = [(35, 42), (41, 45), (49, 33), (59, 62), (76, 75), (80, 96), (85, 91)]

# Anos correspondentes aos pontos de dados, representados como tuplas de strings
anos = [('2014', '2015'), ('2014', '2015'), ('2014', '2015'), ('2014', '2015'),
        ('2014', '2015'), ('2014', '2015'), ('2014', '2015')]

# Plotando cada ponto de dado com seu ano correspondente usando uma linha preta e marcadores circulares
for dado_, ano_ in zip(dados, anos):
    ax.plot(ano_, dado_, color='black', marker='o')

# Removendo as bordas direita, esquerda e superior do gráfico
ax.spines['right'].set_visible(False)  # Remover a borda direita
ax.spines['left'].set_visible(False)   # Remover a borda esquerda
ax.spines['top'].set_visible(False)    # Remover a borda superior

# Removendo os marcadores do eixo y
ax.get_yaxis().set_visible(False)  # Ocultar os marcadores do eixo y

# Definindo o rótulo para o eixo x como 'Ano da pesquisa'
ax.set_xlabel('Ano da pesquisa')  # Definir o rótulo do eixo x

# Adicionando rótulos de porcentagem aos pontos de dados para cada ano
for dado_, ano_ in zip(dados, anos):
    ano_anterior, ano_posterior = ano_
    dado_anterior, dado_posterior = dado_

    # Adicionando texto com a porcentagem à esquerda e à direita de cada ponto de dado
    ax.text(ano_anterior, dado_anterior, f'{dado_anterior}%  ', ha='right')
    ax.text(ano_posterior, dado_posterior, f'  {dado_posterior}%', ha='left')

# Textos adicionais para categorias e descrições
texto = ['Colegas', 'Cultura', 'Ambiente de trabalho', 'Liderança',
         'Desenvolvimento de carreira', 'Gratificações e reconhecimento', 'Gestão de desempenho']

# Adicionando anotações (textos) para cada categoria ao lado do ponto correspondente
for itr in range(len(texto)):
    ax.annotate(texto[itr], xy=('2014', dados[itr][0]), xytext=(-0.35, dados[itr][0]), ha='right')

# Adicionando anotações extras para explicar os rótulos na lateral direita
ax.annotate('Categoria |\nda pesquisa |', xy=('2014', 96), xytext=(-0.35, 96), ha='right')
ax.annotate(' Porcentagem\n Favorável', xy=('2014', 96), xytext=(-0.35, 96), ha='left')

plt.show()  # Exibindo o gráfico
```
