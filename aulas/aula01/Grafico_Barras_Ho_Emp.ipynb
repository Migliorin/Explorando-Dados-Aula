# Bibliotecas

```python
import matplotlib.pyplot as plt # Importando a biblioteca matplotlib para plotagem
import matplotlib.patches as mpatches # Importando recurso para criar legendas customizadas
import seaborn as sns # Importando a biblioteca seaborn para plotagem
```

# Esqueleto do gr√°fico

```python
# Dados para cada dataset (valores percentuais)
dados = [[20, 30, 40, 10],  # Dataset 1
         [10, 35, 25, 30],  # Dataset 2
         [25, 20, 15, 40]]  # Dataset 3

# Cores para cada parte das barras
cores = sns.color_palette()[:4]  # Define uma paleta de cores com base no seaborn para as barras

# Significado de cada cor nas barras
emocoes = ['Smile', 'Fear', 'Disgusting', 'Neutral']  # Nomeia o significado de cada cor nas barras

# Nomes dos Dataset
categorias = ['Dataset 1', 'Dataset 2', 'Dataset 3']  # Nomes dos datasets

fig, ax = plt.subplots(figsize=(7,5))  # Cria uma figura e um eixo com o tamanho especificado

# Desenha as barras horizontais para o primeiro valor de cada dataset com cores diferentes
ax.barh(categorias[2], dados[2][0], color=cores[0])  # Plota a barra para o Dataset 3 com o primeiro valor
ax.barh(categorias[1], dados[1][0], color=cores[0])  # Plota a barra para o Dataset 2 com o primeiro valor
ax.barh(categorias[0], dados[0][0], color=cores[0])  # Plota a barra para o Dataset 1 com o primeiro valor

# Adiciona a legenda para a cor utilizada na barra
ax.legend(handles=[mpatches.Patch(color=cores[0], label=emocoes[0])])  # Cria uma legenda para a cor e sua emo√ß√£o correspondente

plt.show()
```

# Inserindo mais emo√ß√µes

```python
# Dados para cada dataset (valores percentuais)
dados = [[20, 30, 40, 10],  # Dataset 1
         [10, 35, 25, 30],  # Dataset 2
         [25, 20, 15, 40]]  # Dataset 3

# Cores para cada parte das barras
cores = sns.color_palette()[:4]  # Define uma paleta de cores com base no seaborn para as barras

# Significado de cada cor nas barras
emocoes = ['Smile', 'Fear', 'Disgusting', 'Neutral']  # Nomeia o significado de cada cor nas barras

# Nomes dos Dataset
categorias = ['Dataset 1', 'Dataset 2', 'Dataset 3']  # Nomes dos datasets

fig, ax = plt.subplots(figsize=(7, 5))  # Cria uma figura e um eixo com o tamanho especificado

# Desenha as barras horizontais para representar valores cumulativos em diferentes cores
ax.barh(categorias[2], dados[2][0] + dados[2][1] + dados[2][2] + dados[2][3], color=cores[3])  # Barra para o √∫ltimo valor com a cor correspondente
ax.barh(categorias[2], dados[2][0] + dados[2][1] + dados[2][2], color=cores[2])  # Barra para o pen√∫ltimo valor com a cor correspondente
ax.barh(categorias[2], dados[2][0] + dados[2][1], color=cores[1])  # Barra para o antepen√∫ltimo valor com a cor correspondente
ax.barh(categorias[2], dados[2][0], color=cores[0])  # Barra para o primeiro valor com a cor correspondente

ax.barh(categorias[1], dados[1][0], color=cores[0])  # Plota a barra para o Dataset 2 com o primeiro valor
ax.barh(categorias[0], dados[0][0], color=cores[0])  # Plota a barra para o Dataset 1 com o primeiro valor

# Adiciona a legenda para cada cor utilizada nas barras
ax.legend(handles=[
    mpatches.Patch(color=cores[0], label=emocoes[0]),  # Legenda para a primeira cor e emo√ß√£o
    mpatches.Patch(color=cores[1], label=emocoes[1]),  # Legenda para a segunda cor e emo√ß√£o
    mpatches.Patch(color=cores[2], label=emocoes[2]),  # Legenda para a terceira cor e emo√ß√£o
    mpatches.Patch(color=cores[3], label=emocoes[3])   # Legenda para a quarta cor e emo√ß√£o
])  # Cria uma legenda para cada cor e sua emo√ß√£o correspondente

plt.show()  # Exibe o gr√°fico
```

# Compactando o c√≥digo e estilizando

```python
# Dados para cada dataset (valores percentuais)
dados = [[10, 20, 30, 40],  # Dataset 1
         [10, 35, 25, 30],  # Dataset 2
         [15, 25, 20, 40]]  # Dataset 3

# Cores para cada parte das barras
cores = sns.color_palette()[:4]  # Define uma paleta de cores com base no seaborn para as barras

# Significado de cada cor nas barras
emocoes = ['Smile', 'Fear', 'Disgusting', 'Neutral']  # Nomeia o significado de cada cor nas barras

# Nomes dos Dataset
categorias = ['Dataset 1', 'Dataset 2', 'Dataset 3']  # Nomes dos datasets

fig, ax = plt.subplots(figsize=(7, 5))  # Cria uma figura e um eixo com o tamanho especificado

# Desenha as barras horizontais para representar valores cumulativos em diferentes cores
for itr in range(2,-1,-1):
    ax.barh(categorias[itr], dados[itr][0] + dados[itr][1] + dados[itr][2] + dados[itr][3], color=cores[3])  # Barra para o √∫ltimo valor com a cor correspondente
    ax.barh(categorias[itr], dados[itr][0] + dados[itr][1] + dados[itr][2], color=cores[2])  # Barra para o pen√∫ltimo valor com a cor correspondente
    ax.barh(categorias[itr], dados[itr][0] + dados[itr][1], color=cores[1])  # Barra para o antepen√∫ltimo valor com a cor correspondente
    ax.barh(categorias[itr], dados[itr][0], color=cores[0])  # Barra para o primeiro valor com a cor correspondente

# Adiciona a legenda para cada cor utilizada nas barras
ax.legend(handles=[
    mpatches.Patch(color=cores[0], label=emocoes[0]),  # Legenda para a primeira cor e emo√ß√£o
    mpatches.Patch(color=cores[1], label=emocoes[1]),  # Legenda para a segunda cor e emo√ß√£o
    mpatches.Patch(color=cores[2], label=emocoes[2]),  # Legenda para a terceira cor e emo√ß√£o
    mpatches.Patch(color=cores[3], label=emocoes[3])   # Legenda para a quarta cor e emo√ß√£o
],bbox_to_anchor=(1.25, 1.))  # Cria uma legenda para cada cor e sua emo√ß√£o correspondente

# Removendo as bordas direita, esquerda e superior do gr√°fico
ax.spines['right'].set_visible(False)   # Remover a borda direita
ax.spines['left'].set_visible(False)    # Remover a borda esquerda
ax.spines['top'].set_visible(False)     # Remover a borda superior
ax.spines['bottom'].set_visible(False)  # Remover a borda inferior

# Removendo os marcadores do eixo x
ax.get_xaxis().set_visible(False)  # Ocultar os marcadores do eixo x

plt.show()  # Exibe o gr√°fico
```
